Phase 5: ANOVA Test
Hypothesis
H0 (Null Hypothesis):
Company size has no impact on AI adoption.

H1 (Alternative Hypothesis):
Company size significantly impacts AI adoption.

```python
from scipy.stats import f_oneway

startup = df[df['company_size'] == 'Startup']['ai_adoption_rate']
sme = df[df['company_size'] == 'SME']['ai_adoption_rate']
enterprise = df[df['company_size'] == 'Enterprise']['ai_adoption_rate']

f_stat, p_value = f_oneway(startup, sme, enterprise)

print("F-Statistic:", f_stat)
print("P-Value:", p_value)
```

```python
if p_value < 0.05:
    print("Reject H0")
else:
    print("Fail to Reject H0")
```

H0: Industry has no impact on AI adoption.

H1: Industry significantly impacts AI adoption.

# Phase 5: Hypothesis Testing (ANOVA)

## Null Hypothesis (H0)
Company size has no impact on AI adoption.

## Alternative Hypothesis (H1)
Company size significantly impacts AI adoption.

## Test Used
One-Way ANOVA

## Results
F-Statistic = 3746.05

P-Value = 0.0

## Decision
Since the p-value is less than 0.05, the null hypothesis is rejected.

## Conclusion
Company size significantly impacts AI adoption.

## Business Interpretation
The analysis shows that AI adoption differs significantly among Startups, SMEs, and Enterprises. Larger organizations generally have greater financial resources, technical expertise, and infrastructure to implement AI solutions effectively. Therefore, company size is an important factor influencing AI adoption strategies and outcomes.

Phase 6: Organization Segmentation
Create Segments

```python
def classify_org(row):

    if row['ai_adoption_rate'] >= 80:
        return 'AI Leader'

    elif row['ai_adoption_rate'] >= 60:
        return 'Early Adopter'

    elif row['employee_satisfaction_score'] >= 85:
        return 'High Satisfaction'

    elif row['cost_reduction_percent'] >= 25:
        return 'High ROI'

    else:
        return 'Slow Adopter'

df['organization_segment'] = df.apply(classify_org, axis=1)

df['organization_segment'].value_counts()
```

```python
import matplotlib.pyplot as plt

df['organization_segment'].value_counts().plot(
    kind='bar',
    figsize=(8,5)
)

plt.title('Organization Segments')
plt.xlabel('Segment')
plt.ylabel('Count')
plt.show()
```
